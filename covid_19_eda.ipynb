```python
df = pd.read_csv('/content/compact.csv')
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
countries = ['India', 'United States', 'Brazil', 'Russia', 'United Kingdom']
df_subset = df[df['country'].isin(countries)]
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 10), sharex=True, sharey=True)
axes = axes.flatten()

for i, country in enumerate(countries):
    country_data = df_subset[df_subset['country'] == country]
    axes[i].plot(country_data['date'], country_data['new_cases'], label=country, color='blue')
    axes[i].set_title(country)
    axes[i].tick_params(axis='x', rotation=45)
if len(countries) % 2 != 0:
    fig.delaxes(axes[-1])

plt.tight_layout()
plt.suptitle('ðŸ“ˆ Daily New COVID-19 Cases by Country (Matplotlib)', fontsize=16, y=1.02)
plt.show()
```
